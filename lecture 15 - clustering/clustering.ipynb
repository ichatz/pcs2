```python
url = 'http://aima.cs.berkeley.edu/data/iris.csv'

import csv
f = open("iris.csv")
rawdata = []
for row in csv.reader(f, delimiter=','):
    rawdata.append(row)
```

```python
for row in rawdata:
    for value in range(0, len(row)-1):
        row[value] = float(row[value])
```

```python
import matplotlib.pyplot as plt
plt.plot([row[0] for row in rawdata if row[4] == 'setosa'], 
        [row[1] for row in rawdata if row[4] == 'setosa'],
        'o', label='setosa')

plt.plot([row[0] for row in rawdata if row[4] == 'versicolor'], 
        [row[1] for row in rawdata if row[4] == 'versicolor'],
        'o', label='versicolor')

plt.plot([row[0] for row in rawdata if row[4] == 'virginica'], 
        [row[1] for row in rawdata if row[4] == 'virginica'],
        'o', label='virginica')

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
plt.show()
```

```python
columnTitles = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"]
for column in range(0,4):
    xmin = min([row[column] for row in rawdata])
    xmax = max([row[column] for row in rawdata])
    plt.figure()
    plt.subplot(311)
    plt.title(columnTitles[column])
    plt.xlim(xmin, xmax)
    plt.hist([row[column] for row in rawdata if row[4] == 'setosa'])
    plt.subplot(312)
    plt.xlim(xmin, xmax)
    plt.hist([row[column] for row in rawdata if row[4] == 'versicolor'])
    plt.subplot(313)
    plt.xlim(xmin, xmax)
    plt.hist([row[column] for row in rawdata if row[4] == 'virginica'])
    plt.savefig("plot-" + str(column) + ".png")
    plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
plt.clf()
fig = plt.figure(1, figsize=(7, 4))
ax = Axes3D(fig, elev=-150, azim=110)

categories = ['setosa', 'versicolor', 'virginica']
color = ['b','g','r']

for cat in range(0,3):
    ax.scatter3D([row[0] for row in rawdata if row[4]==categories[cat]],
                [row[1] for row in rawdata if row[4]==categories[cat]],
                [row[2] for row in rawdata if row[4]==categories[cat]],
                color=color[cat])

plt.title("Sepal Width vs Sepal Length vs Petal Length")
plt.show()
```
